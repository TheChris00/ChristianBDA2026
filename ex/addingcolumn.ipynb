```python
import pandas as pd


df = pd.read_csv("/Users/chris/Desktop/ChristianBDA2026/data/symbol_info_3-25.csv")

df_sample = df.sample(30, random_state = 43)
df_sample = df_sample[['market_cap', 'sector']]


total_marketcap = df_sample['market_cap'].sum() 

percentege_marketcap = (df_sample['market_cap']/ total_marketcap) * 100

# now adding "percentege_marketcap" as new column

df_sample['market_cap%'] = (df_sample['market_cap']/ total_marketcap) * 100

print(df_sample)

```

```python
# apply and lambda

df_sample['market_cap%'] = df_sample['market_cap'].apply(lambda x: (x / total_marketcap))
print(df_sample)

```
