### Exercise 1.5 — Add a computed column
Compute a new column called `market_cap_B` that expresses `market_cap` in **billions** (divide by 1,000,000,000).
Then compute another column `market_cap_%` that is each company's share of the **total** market cap in the sample (as a percentage).

> **Think about it:** the professor used `apply(lambda x: ...)` as an alternative. Write the same computation using `apply()` and verify you get the same result.


```python

df_sample['market_cap_B'] = df_sample['market_cap'].apply(lambda x: x / 1000000000)


total_market_value = df_sample['market_cap'].sum()

df_sample['market_cap_%']= df_sample['market_cap'].apply(lambda x: (x / total_market_value) * 100)

print(df_sample)

```

### Exercise 1.6 — GroupBy and aggregation

### Exercise 1.7 — Set index and multi-index

```python
df_index = df_sample.set_index(['sector', 'symbol'])

df_index_sorted = df_index.sort_values(
  by = ['sector', 'symbol'],
  ascending=[True, False]
  
  )

print(df_index_sorted.loc['Technology'])




'''

Perché l'ordine giusto è ['sector', 'symbol']
                                               
  Questa è la domanda più importante. Non è solo convenzione — riflette la
  gerarchia reale dei dati.                                                     
   
  Guarda la relazione tra settore e simbolo:                                    
  - Financial Services → CFR, TD, ICE, ALL, COIN, ONB (tanti simboli)
  - AAPL → appartiene a un solo settore                                         
                                       
  Un settore contiene molti simboli. Un simbolo appartiene a un solo settore.   
  Quindi:                                                                       
                                                                                
  sector  (livello esterno)                                                     
    └── symbol  (livello interno)

  È come un libro: prima il capitolo (sector), poi la pagina (symbol). Non ha   
  senso organizzarlo al contrario.

'''






'''

Bonus: sort_values vs sort_index
                                                                                
  Stai usando sort_values(by=['sector', 'symbol']) su un DataFrame dove sector e
   symbol sono l'indice, non colonne. Il metodo corretto per ordinare per i     
  livelli di un MultiIndex è sort_index():
                                                                                
  df_index_sorted = df_index.sort_index(ascending=[True, False])
  #                                                 ↑ sector  ↑ symbol          
  #                                              A-Z       Z-A
                                                                                
  Quindi la cella corretta diventa:                                             
                                                                                
  df_index = df_sample.set_index(['sector', 'symbol'])                          
                  
  df_index_sorted = df_index.sort_index(ascending=[True, False])                
  
  # Tutti i titoli Technology                                                   
  print(df_index_sorted.loc['Technology'])

  # Simbolo specifico
  print(df_index_sorted.loc[('Technology', 'APP')])

'''
```

### Exercise 1.8 — Filtering with boolean masks
Starting from the **full** dataset (all 1,000 rows loaded with `pd.read_csv`):
- Keep only companies where `market_cap` is **above the median**
- From those, keep only companies in the **Healthcare** or **Technology** sector
- Print how many companies remain and the top 5 by market cap

> **Think about it:** what is the difference between `&` and `|` in boolean filtering? Write a version of the filter using `|` and explain what it returns.


```python
df_all = pd.read_csv('/Users/chris/Desktop/ChristianBDA2026/data/symbol_info_3-25.csv')

print('---Above the Median---')
market_cap = df_all['market_cap'] > df_all['market_cap'].median()

print('---Healthcare and Technology---')
market_cap_sector = df_all['sector'].isin(['Healthcare', 'Technology'])

df_filtered = df_all[market_cap & market_cap_sector]

print(f"Companies: {len(df_filtered)}")
print(df_filtered.nlargest(5, 'market_cap')[['symbol', 'sector', 'market_cap']])
```

```python
#df_sector_unique = df_all.unique('sector')

'''
  # ❌ unique() non esiste sul DataFrame intero
  df_all.unique('sector')                                                       
                  
  # ✅ unique() è un metodo della Series — prima selezioni la colonna           
  df_all['sector'].unique()
                                                                                
  La regola da ricordare: unique(), value_counts(), nunique() si chiamano sempre
   sulla colonna, non sul DataFrame.
'''
##df_sector_unique = df_all['sector'].unique()
##print(df_sector_unique.value_counts())

'''
 Problema: stai lavorando su df_sector_unique invece che su df_all             
                                                                                
  .unique() restituisce un numpy array semplice (solo i valori unici), non un
  DataFrame. Non ha colonne, non puoi aggiungerci sector_upper.                 
                  
  df_sector_unique = df_all['sector'].unique()                                  
  # → array(['Financial Services', 'Energy', 'Technology', ...])
  # è solo una lista di nomi, non ha struttura DataFrame     

'''
print(df_all['sector'].unique())
print(df_all['sector'].value_counts())

df_all['sector_upper'] = df_all['sector'].str.upper()

print('-- Row that contains Consumer---')

##df_consumer = df_all['sector'].str.contains('Consumer', case = False)
##print(df_consumer)

'''
  Immagina un foglio Excel con 1000 righe
                                                                                
  Hai df_all con 1000 aziende. Vuoi tenere solo quelle del settore "Consumer".
                                                                                
  In Excel lo faresti manualmente: scorri le righe, segni quelle giuste, le     
  copi. Pandas fa la stessa cosa ma in automatico — in due passi.               
                                                                                
  ---             
  Passo 1 — Crei la maschera (il "segnaposto")
                                                                                
  df_consumer = df_all['sector'].str.contains('Consumer')
                                                                                
  Questo non filtra niente. Crea solo una colonna di True/False — una per ogni  
  riga:                                                                         
                                                                                
  riga 0    →  False   (Technology)
  riga 1    →  False   (Financial Services)
  riga 2    →  True    (Consumer Cyclical)   ← contiene "Consumer"
  riga 3    →  True    (Consumer Defensive)  ← contiene "Consumer"              
  riga 4    →  False   (Healthcare)                                             
  ...                                                                           
                                                                                
  Questo è la maschera — 1000 valori True/False. Se stampi df_consumer vedi     
  proprio questa colonna di True/False, non le righe vere.
                                                                                
  ---             
  Passo 2 — Usi la maschera per filtrare
                                                                                
  df_all[df_consumer]
                                                                                
  Pandas legge: "dammi solo le righe dove la maschera è True". È come dire      
  "tienimi solo le righe segnate".
                                                                                
  ---             
  Riassunto visivo

  df_all          df_consumer      df_all[df_consumer]
  ----------      -----------      -------------------                          
  Technology   →   False      →   (scartata)                                    
  Financial    →   False      →   (scartata)                                    
  Consumer C.  →   True       →   Consumer Cyclical  ✅                         
  Consumer D.  →   True       →   Consumer Defensive ✅
  Healthcare   →   False      →   (scartata)            



   Il pattern è sempre lo stesso in pandas:
  1. Crei la condizione → ottieni True/False                                    
  2. La metti dentro df[] → ottieni le righe filtrate                        
                  
  ---       
'''

df_consumer = df_all['sector'].str.contains('Consumer', case = False)
print(df_all[df_consumer]) # we using the mask to filtring


print('--Replacingn---')



#df_all['Sector_Replace'] = df_all['sector'].replace([{'Financial_Servoces': 'Finance'}, {'Consumer_Cyclical': 'Cyclical'}])
#print(df_all)

''' 

 Quasi! Due errori nel replace:                                                
                                                     
  1. Lista di dizionari invece di un dizionario solo                            
                                                                                
  # ❌ tuo — lista con due dizionari separati                                   
  .replace([{'Financial_Servoces': 'Finance'}, {'Consumer_Cyclical':            
  'Cyclical'}])                                                                 
                                                                                
  # ✅ corretto — un solo dizionario con due coppie dentro                      
  .replace({'Financial Services': 'Finance', 'Consumer Cyclical': 'Cyclical'})
                                                                                
  2. Typo nelle chiavi — underscore _ invece di spazio, e "Servoces" invece di  
  "Services"                                                                    
                                                                                
  'Financial_Servoces'  # ❌
  'Financial Services'  # ✅
                                                                                
  'Consumer_Cyclical'   # ❌
  'Consumer Cyclical'   # ✅                                                    
                  
  I nomi devono corrispondere esattamente a quelli nel dataset — spazi inclusi. 
   
  Codice corretto:                                                              
                  
  df_all['sector'] = df_all['sector'].replace({                                 
      'Financial Services': 'Finance',
      'Consumer Cyclical':  'Cyclical'
  })
  print(df_all['sector'].value_counts())

'''


df_all['sector'] = df_all['sector'].replace({'Financial Services': 'Finance', 'Consumer Cyclical': 'Cyclical'})
print(df_all['sector'].value_counts())
```
