```python
import pandas as pd

df = pd.read_csv('/Users/chris/Desktop/ChristianBDA2026/data/symbol_info_3-25.csv')

df_sample = df.sample(50, random_state = 40)
df_sample = df_sample[['market_cap', 'sector']]
```

# GROUPING

## Step 1: Group the rows by 'sector'

```python

grouped = df_sample.groupby('sector')   # divide il df in 8 gruppi(uno per settore)
                                        # Crea un oggetto "DataFrameGroupBy" — come un contenitore

print("Step 1 - type:", type(grouped).__name__)
print('Number of groups:', grouped.ngroups) #`.ngroups` = proprietà che dice quanti gruppi ci sono
print('Group sizes:')
print(grouped.size()) #.size() = conta quante righe in ogni gruppo
```

### PEEK INSIDE

```python
print('Let peak inside one group, Financial Services :')

print(grouped.get_group('Financial Services')) # get_group('nome') = prende UN gruppo specifico

                                                # Ti mostra tutte le righe di quel settore
```

### Step 2: Select the 'market_cap' column from each group

```python
market_cap_by_sector = grouped['market_cap'] # Da ogni gruppo, prendi SOLO la colonna market_cap

print('Step 2 - now we only keep the market_cap values per groups:')
for name, values in market_cap_by_sector:
    print(f' {name}: {values.tolist()}')
```

# Step 3: Apply summary functions — sum, mean, and count

```python
df_grouped = market_cap_by_sector.agg([ 'sum', 'mean', 'count'])
print(df_grouped)
```

```python
# othe method

df_grouped = df_sample.groupby('sector')['market_cap'].agg(['sum', 'mean', 'count']).reset_index()

print(df_grouped)
```
